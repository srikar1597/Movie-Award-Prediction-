# 11.Feature Scaling for Train and Test dataset

```python
#importing the Standardscaler package for feature Scaling
#transforming x_train 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_std=sc.fit_transform(X_train)
X_train_std
```

```python
#converting the X_train_std into dataframe
num_scaled_df = pd.DataFrame(X_train_std, columns=x.columns)
num_scaled_df.head()
```

# 12.Applying the Classification Algorithms

# Classification Algorithms

1.Logistic Regression

2.Naive Bayes

```python
# Importing the all Classification algorithms libraries
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

# SVM Classifier

```python
# Applying the SVM Classifier to training data

clf_svm_l =SVC(kernel='linear', C=100)
clf_svm_l.fit(X_train_std, y_train)
```

```python
# Predicting the Values from trainimg and testing

#y_train_pred = clf_svm_l.predict(X_train_std)
y_test_pred = clf_svm_l.predict(X_test_std)
y_test_pred
```

```python
# Checking the acuuracy score for svm algorithm

accuracy_score(y_test, y_test_pred)
```

# Model Performance SVM classifier

# KNeighborsClassifier 

```python
NC=KNeighborsClassifier()

NC.fit(X_train_std, y_train)
y_test_pred = NC.predict(X_test_std)
accuracy_score(y_test, y_test_pred)
```

# LogisticRegression

```python
LR=LogisticRegression()
LR.fit(X_train_std, y_train)
y_test_pred = LR.predict(X_test_std)
accuracy_score(y_test, y_test_pred)
```

# GaussianNB

```python
GNB=GaussianNB()
GNB.fit(X_train_std,y_train)
y_test_pred=GNB.predict(X_test_std)
accuracy_score(y_test,y_test_pred)
```

# DecisionTreeClassifier

```python
DT=DecisionTreeClassifier()
DT.fit(X_train_std, y_train)
y_test_pred = DT.predict(X_test_std)
accuracy_score(y_test, y_test_pred)
```

# LinearDiscriminantAnalysis

```python
LK=LinearDiscriminantAnalysis()
LK.fit(X_train_std, y_train)
y_test_pred = LK.predict(X_test_std)
accuracy_score(y_test, y_test_pred)
```

# RandomForestClassifier

```python
RT=RandomForestClassifier()
RT.fit(X_train_std, y_train)
y_test_pred = RT.predict(X_test_std)
accuracy_score(y_test, y_test_pred)
```
